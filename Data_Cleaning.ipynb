```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as plt
from matplotlib import pyplot
import seaborn as sns
%matplotlib inline
```

```
path =  "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df = pd.read_csv(path, names=headers)
```

```
df.head()
```

# Handling missing values

```
df.replace("?", np.nan, inplace=True)
missing_data = df.isnull()
print(missing_data.sum())
```

Im going to replace the numerical values with the mean of the column, except for the price missing rows which im going to drop, the categorical are going to be replaced with the most frequent value.


```
df.head(3)
```

```
#DONT RUN THIS LINE 
median_normalizedlosses = df["normalized-losses"].median()
df["normalized-losses"].fillna(median_normalizedlosses, inplace=True)
median_bore = df["bore"].median()
df["bore"].fillna(median_bore, inplace=True)
median_stroke = df["stroke"].median()
df["stroke"].fillna(median_stroke, inplace=True)
median_peakrpm = df["peak-rpm"].median()
df["peak-rpm"].fillna(median_peakrpm, inplace=True)
median_horsepower = df["horsepower"].median()
df["horsepower"].fillna(median_horsepower, inplace=True)
```

```
df.dropna(subset=["price"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df["num-of-doors"].fillna(df["num-of-doors"].value_counts().idxmax(), inplace=True)
```

```
print("Total missing values:", df.isnull().values.sum())
```

All missing values are now handled

# Formatting the incorrect data types

Due to the fact that many variables have incorrect data types which will affect the model, they need to be handled.

```
df.dtypes
```

Notice that the variables horsepower, price, peak-rpm, bore, stroke and normalized losses are all object types, which doesn't makes sense.

```
df.head(5)
```

```
# DONT TOUCH THIS LINE TWICE
#Changing the datatypes
df[["peak-rpm", "normalized-losses", "horsepower"]] = df[["peak-rpm", "normalized-losses", "horsepower"]].astype("int")
df[["bore", "stroke", "price"]] = df[["bore", "stroke", "price"]].astype("float")
```

```
df.dtypes
```

# Scaling the data

Since the many alike variables have very different scales, they need to be scaled for proper analysis, for that, im going to use the min max scaling method.

```
df.head()
```

You can see that the variables: length, width, height, are in very different scales.

```
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[["length", "width", "height"]] = scaler.fit_transform(df[["length", "width", "height"]])
```

```
df[["length", "width", "height"]].head()
```

## Handling categorical variables

Im going to do this using dummy variables

```
df.select_dtypes(include=object).head(5)
```

Im only going to deal with aspiration, num of doors, fuel type and engine location 

```
dummy_aspiration = pd.get_dummies(df["aspiration"])
dummy_aspiration.rename(columns={"std":"aspiration-std", "turbo":"aspiration-turbo"}, inplace=True)
df = pd.concat([df, dummy_aspiration], axis=1)
df.drop("aspiration", axis=1, inplace=True)

dummy_numofdoors = pd.get_dummies(df["num-of-doors"])
dummy_numofdoors.rename(columns={"four":"nofdoors-four", "two":"nofdoors-two"}, inplace=True)
df = pd.concat([df, dummy_numofdoors], axis=1)
df.drop("two", axis=1, inplace=True)

dummy_fueltype = pd.get_dummies(df["fuel-type"])
dummy_fueltype.rename(columns={'fuel-type-diesel':'gas', 'fuel-type-diesel':'diesel'}, inplace=True)
df = pd.concat([df, dummy_fueltype], axis=1)
df.drop("fuel-type", axis=1, inplace=True)

dummy_engineloc = pd.get_dummies(df["engine-location"])
dummy_engineloc.rename(columns={ "front": "engine_location-front", "rear": "engine_location-rear"}, inplace=True)
df = pd.concat([df, dummy_engineloc], axis=1)
df.drop("engine-location", axis=1, inplace=True)
df.head(5)
```

```
df.to_csv("clean_carprice.csv")
```
